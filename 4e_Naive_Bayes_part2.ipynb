## Naive Bayes


```python
import pandas as pd
import numpy as np

from sklearn.naive_bayes import MultinomialNB, GaussianNB
from sklearn import preprocessing
```

### Tennis example (multiple predictors)

```python
multi_mixed = """
  Weather
  Temp
  Play
  Sunny
  Hot
  No
  Sunny
  Hot
  No
  Overcast
  Hot
  Yes
  Rainy
  Mild
  Yes
  Rainy
  Cool
  Yes
  Rainy
  Cool
  No
  Overcast
  Cool
  Yes
  Sunny
  Mild
  No
  Sunny
  Cool
  Yes
  Rainy
  Mild
  Yes
  Sunny
  Mild
  Yes
  Rainy
  Mild
  No
  Overcast
  Hot
  Yes

""".split()
print(multi_mixed)
```

```python
weather = multi_mixed[0::3]
temp =  multi_mixed[1::3]
play =  multi_mixed[2::3]

print(weather)
print(temp)
print(play)
```

```python
import pprint
pp = pprint.PrettyPrinter(depth=2, compact=True, width=150)
#pp = PrettyPrint (depth=2, compact=True, width=150) <- This are just different ways to display code in a 'pretty' way.

table = { weather.pop(0): weather,
          temp.pop(0): temp,
          play.pop(0): play
}
pp.pprint(table)
```

```python
# Create labelEncoder
le = preprocessing.LabelEncoder()

# Convert string labels into numbers. Note: this isn't the best way to encode
# weather but we will talk about other encoding methods later on
weather_encoded = le.fit_transform(weather)
temp_encoded = le.fit_transform(temp)
label_encoded = le.fit_transform(play)

print('Label:', label_encoded)
print('Temp:', temp_encoded)
print('Weather:', weather_encoded)
```

Show encoding.

```python
print(sorted(set(zip(weather_encoded, weather))))
print(sorted(set(zip(temp_encoded, temp))))
print(sorted(set(zip(label_encoded, play))))
```

```python
#Combine weather and temp into data frame
my_dict = {'Weather': weather_encoded, 'Temp': temp_encoded}
features = pd.DataFrame(my_dict)
features
```

```python
features.iloc[0].to_dict()
```

```python
features.iloc[0:5].to_dict(orient= "records")
```

```python
# Create a Naive Bayes Classifier
model = MultinomialNB()

# Train the model using the training sets
model.fit(features,label_encoded) ;
```

```python
# Predict Output
input = pd.DataFrame({'Weather': [1], 'Temp': [2]}) # 1: Rainy, 2: Mild
predicted = model.predict( input )
probability = model.predict_proba( input )

print( f"Predicted Value: ", ("No","Yes")[predicted[0]] )
print( f"Probability Yes: {probability[0][1]*100:0.2f}%, No: {probability[0][0]*100:0.2f}%" )
```
