## Python's core components:  



### **1. Fundamental Building Blocks**  


- **Statements** - Instructions that the interpreter executes (e.g., `if`, `for`, `while`, `import`).  
- **Expressions, Conditional Expressions** - Combinations of values, variables, and operators that evaluate to a single value (e.g., `2 + 3`, `x * y`, `x == 5`, `y is None`).  
- **Variables** - Named references to objects in memory (e.g., `x = 10`).  
- **Objects** - Everything in Python is an object, including numbers, functions, and classes.  
- **Values** - Immutable data like `5`, `"hello"`, or `None`.

### **2. Data Structures**  


- **Primitive Data Types** - Integers (`int`), floating-point numbers (`float`), strings (`str`), booleans (`bool`), etc.  
- **Collections** - Lists (`list`), tuples (`tuple`), dictionaries (`dict`), sets (`set`).  
- **Iterators & Generators** - Objects that yield values one at a time (`iter()`, generator functions).  



### **3. Control Flow Constructs**  


- **Branching Statements** - `if`, `elif`, `else`, `match-case`.  
- **Loops** - `for`, `while`.  
- **Exception Handling** - `try`, `except`, `finally`, `raise`.  



### **4. Functions & Modules**  


- **Functions** - Defined using `def`, lambda functions.  
- **Modules & Packages** - Files containing Python code (`import math`), organized into packages (directories with `__init__.py`).  



### **5. Object-Oriented Features**  


- **Classes & Instances** - Defined using `class`, with `__init__` for initialization.  
- **Methods & Attributes** - Functions and data attached to objects.  
- **Inheritance & Polymorphism** - Mechanisms for extending behavior.  



### **6. Special Features**  


- **Decorators** - Functions that modify other functions (`@interact`, `@app.get`).  
- **Metaclasses** - Control class creation.  
- **Context Managers** - `with open('file.txt') as f:`.  
- **Type Hints** - Optional type annotations (`def add(x: int, y: int) -> int:`).  



### **7. Execution Model & Memory Management**  


- **Scopes & Namespaces** - `global`, `local`, `nonlocal`.  
- **Garbage Collection** - Automatic memory management via reference counting and garbage collection.  
- **Bytecode & Virtual Machine** - Python code is compiled to bytecode and executed by the Python Virtual Machine (PVM).
