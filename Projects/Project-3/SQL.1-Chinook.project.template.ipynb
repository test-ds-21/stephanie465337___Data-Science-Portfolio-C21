###Data Questions

```python
with db.connect('chinook.db') as db_con:

  query = '''
    SELECT name
    FROM sqlite_master
    WHERE type = 'table';
  '''

  tables = pd.read_sql_query(query, db_con)

  for table in tables ['name']:

    query = f'''
      SELECT COUNT(1)
      FROM "{table}";
    '''
    row_count = pd.read_sql_query(query, db_con).iloc[0,0]

    query = f'''
      PRAGMA table_info("{table}");
    '''
    columns = pd.read_sql_query(query, db_con)

    print('=' * 60)
    print(f'Table: {table}')
    print('=' * 60)
    print(f'Rows: {row_count}')
    print(f'Columns: {len(columns)}')
    print()

    summary = []

    for _, col in columns.iterrows():

      column = col['name']
      dtype = col['type']

      identifier = 'YES' if col['pk'] == 1 else 'NO'

      query = f'''
        SELECT COUNT(1)
        FROM "{table}"
        WHERE "{column}" IS NULL;
      '''
      nulls = pd.read_sql_query(query, db_con).iloc[0,0]

      query = f'''
        SELECT COUNT(DISTINCT "{column}")
        FROM "{table}";
      '''
      distinct = pd.read_sql_query(query, db_con).iloc[0,0]

      summary.append({
        'Column': column,
        'Type': dtype,
        'Identifier': identifier,
        'Nulls': nulls,
        'Distinct': distinct
      })

    summary_df = pd.DataFrame(summary)

    print(summary_df.to_string(index=False))
    print('\n')
```

```python
#Opens SQLite database so Python can send the SQL queries to it and close the connection when finished.
with db.connect('chinook.db') as db_con:

  #Tables
  # sqlite_master is built into every SQLite db. (Table of Contents)
  # SELECTs the name of the column
  # FROM the catalog
  # WHERE only returns the tables (not indexes or views)?
  query = '''
    SELECT name
    FROM sqlite_master
    WHERE type = 'table';
  '''

  tables = pd.read_sql_query(query, db_con)

  #Loop through each table
  for table in tables ['name']:

    #Rows
    # .iloc[0,0] is to extract the actual value instead of returning the DataFrame.
    #Use "" around {}
    query = f'''
      SELECT COUNT(1)
      FROM "{table}";
    '''
    row_count = pd.read_sql_query(query, db_con).iloc[0,0]

    #Columns
    #PRAGMA table_info() is a SQLite command that tells you everything about the columns.
    #Use "" around {}
    query = f'''
      PRAGMA table_info("{table}");
    '''
    columns = pd.read_sql_query(query, db_con)

    print('=' * 60) #Prints top border
    print(f'Table: {table}') #Prints table name
    print('=' * 60) #Prints bottom border
    print(f'Rows: {row_count}') #Labels row count
    print(f'Columns: {len(columns)}') #Labels column count
    print()

    #This will store a dict for every column.
    #The list will turn into a DataFrame at the end.
    summary = []

    #Loop through every column
    #Each row has 2 pieces of info (index, row). _ removes the index.
    for _, col in columns.iterrows():

      #Pulls the name of the column
      column = col['name']
      #Pulls the dtype (referred to as type) ex INTEGER
      dtype = col['type']

      #Identifiers
      #If pk = 1 it is a primary key (pk comes from SQLite/DO NOT change)
      identifier = 'YES' if col['pk'] == 1 else 'NO'

      #NULLs
      #Use "" around {}
      query = f'''
        SELECT COUNT(1)
        FROM "{table}"
        WHERE "{column}" IS NULL;
      '''
      nulls = pd.read_sql_query(query, db_con).iloc[0,0]

      #Distinct values
      #Produces a value to show how many unique values are in the column.
      #Could be an indicator of a categorical
      #Use "" around {}
      query = f'''
        SELECT COUNT(DISTINCT "{column}")
        FROM "{table}";
      '''
      distinct = pd.read_sql_query(query, db_con).iloc[0,0]

      #Creates a dictionary to store everything.
      summary.append({
        'Column': column,
        'Type': dtype,
        'Identifier': identifier,
        'Nulls': nulls,
        'Distinct': distinct
      })

    #Turns the summary list from above into a DataFrame
    summary_df = pd.DataFrame(summary)

    #Using to_string(index=False) prints each table immediate
    print(summary_df.to_string(index=False)) #index=False removes the added row numbers from Pandas
    print('\n') #Enters the line for space
```

```python
#How many Invoices were there in 2009 and 2011?
#What are the respective total sales for each of those years?

#STRFTIME (String Format Time) Year-Month-Day Hour:Minute:Second
#SQL command to pull something out of a date.
#All options: %Y/4-digit year xxxx, %m/Month xx, %d/Day xx
#Q1: COUNT and label the total amount of invoices in the year.
#Q2: ROUND the total sales in the a year to 2 decimal places
#WHERE keeps the specified years 2009 and 2011
#GROUP BY keeps the total for each year separate
query = '''
  SELECT
    STRFTIME('%Y', InvoiceDate) AS Year,
    COUNT(*) AS NumberOfInvoices,
    ROUND(SUM(Total), 2) AS TotalSales
  FROM invoices
  WHERE STRFTIME('%Y', InvoiceDate) IN ('2009', '2011')
  GROUP BY STRFTIME('%Y', InvoiceDate)
  ORDER BY Year;
'''

with db.connect("chinook.db") as db_con:
  df = pd.read_sql_query(query , db_con)
df
```

```python
import matplotlib.pyplot as plt
```

```python
query = '''
  SELECT
    STRFTIME('%Y', InvoiceDate) AS Year,
    ROUND(SUM(Total),2) AS Sales
  FROM Invoices
  GROUP BY Year
  ORDER BY Year;
'''

with db.connect("chinook.db") as db_con:
  df = pd.read_sql_query(query , db_con)

plt.plot(df["Year"], df["Sales"], marker='o')
plt.xticks(rotation=45)
plt.title("Total Sales by Year")
plt.xlabel("Year")
plt.ylabel("Sales ($)")
plt.grid(True)
plt.savefig('sales_by_year.png')
plt.show()
```
