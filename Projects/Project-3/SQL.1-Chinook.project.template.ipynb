## Phase 2. Map the Terrain (Get Familiar with the Data)



Data never exists in a vacuum. By writing queries, you will uncover the shape, structure, and quirks of the Chinook dataset. Again, using the SQL statements below, you'll learn how the tables relate to one another, what a single row represents in different contexts, and where specific types of information live.  Here's where you should get familiar with the ERD and join tables in various ways.



```python
summary = []

with db.connect('chinook.db') as db_con:

  #Tables
  query = '''
    SELECT name
    FROM sqlite_master
    WHERE type='table';
  '''

  tables = pd.read_sql_query(query, db_con)

  #Loop through each table
  for table in tables['name']:

    #Number of rows
    query = f'''
      SELECT COUNT(*) AS Rows
      FROM '{table}';
    '''
    rows = pd.read_sql_query(query, db_con).iloc[0,0]

    #Number of columns
    query = f'''
      PRAGMA table_info('{table}');
    '''
    columns = pd.read_sql_query(query, db_con)

    #Loop through each column
    for _, col in columns.iterrows():

      column = col['name']
      dtype = col['type']
      pk = 'Yes' if col['pk'] == 1 else 'No'

      #NULL values
      query = f'''
        SELECT CoUNT(*)
        FROM '{table}'
        WHERE '{column}' IS NULL;
      '''
      nulls = pd.read_sql_query(query, db_con).iloc[0,0]

      #Identifiers
      query = f'''
        SELECT COUNT(DISTINCT '{column}')
        FROM '{table}';
      '''
      distinct = pd.read_sql_query(query, db_con).iloc[0,0]

      #Classify each column
      if pk == 'YES':
        category = 'Identifier'
      elif 'CHAR' in dtype.upper() or 'TEXT' in dtype.upper():
        category = 'Categorical'
      elif 'REAL' in dtype.upper() or 'FLOAT' in dtype.upper():
        category = 'Continuous'
      elif 'INT' in dtype.upper():
        category = 'Discrete'
      else:
        category = 'Other'

      summary.append({
          'Table': table,
          'Rows': rows,
          'Column': column,
          'Type': dtype,
          'Identifier': pk,
          'Nulls': nulls,
          'Distinct': distinct,
          'Category': category
      })

summary_df = pd.dataframe(summary)

summary_df
```
