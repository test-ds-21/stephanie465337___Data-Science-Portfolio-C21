### Basic Queries


#### SELECT (with * and with column names)


```python
query = '''
  SELECT *
  FROM employees
'''

with db.connect("chinook.db") as db_con:
  df = pd.read_sql_query( query , db_con )
df
```

#### WHERE


```python
query = '''
  SELECT *
  FROM employees
  WHERE employeeid = 3
'''

with db.connect("chinook.db") as db_con:
  df = pd.read_sql_query( query , db_con )
df
```

#### AND


```python
query = '''
  SELECT *
  FROM employees
  WHERE employeeid > 3
    AND State = 'AB'
'''

with db.connect("chinook.db") as db_con:
  df = pd.read_sql_query( query , db_con )
df
```

#### OR


```python
query = '''
  SELECT *
  FROM employees
  WHERE City = 'Calgary'
    OR City = 'Lethbridge'
'''

with db.connect("chinook.db") as db_con:
  df = pd.read_sql_query( query , db_con )
df
```

#### LIKE (with % or _ wildcard)


```python
query = '''
  SELECT *
  FROM employees
  WHERE FirstName LIKE 'M%'
'''

with db.connect("chinook.db") as db_con:
  df = pd.read_sql_query( query , db_con )
df
```

#### BETWEEN


```python
query = '''
  SELECT *
  FROM employees
  WHERE EmployeeId BETWEEN 4 AND 7
'''

with db.connect("chinook.db") as db_con:
  df = pd.read_sql_query( query , db_con )
df
```

#### LIMIT



```python
query = '''
  SELECT *
  FROM employees
  WHERE City LIKE 'C%'
  LIMIT 3
'''

with db.connect("chinook.db") as db_con:
  df = pd.read_sql_query( query , db_con )
df
```

```python
query = '''
  SELECT *
  FROM playlists AS p1
  INNER JOIN playlist_track AS p2
  ON p1.PlaylistId = p2.PlaylistId
  WHERE Name = 'Music'
    AND p1.PlaylistId = 1
'''

with db.connect("chinook.db") as db_con:
  df = pd.read_sql_query( query , db_con )
df
```
