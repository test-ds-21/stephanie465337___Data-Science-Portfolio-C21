Build each of the following games in Python:

## Rock, Paper, Scissors

```python
import random
import ipywidgets as widgets

from IPython.display import display

#List of choices
choices =['Rock', 'Paper', 'Scissors']

#Dictionary showing what beats what
wins = {
    'Rock':'Scissors',
    'Paper':'Rock',
    'Scissors':'Paper'
}

#Create dropdown
user = widgets.Dropdown(
    options=choices,
    value='Rock',
    description='Rock, paper or scissors?',
    style = {'description_width': 'initial'}, #Shows entire description
    layout=widgets.Layout(width='250px')
)

#Create button
button = widgets.Button(
    description='Shoot',
    button_style='success' #Makes button green
)

#Creates area where results will be printed
output = widgets.Output()

#Create function to play game
def play_game(b): # b represents button
  with output:
    output.clear_output()

    #Random choice
    computer = random.choice(choices)
    player = user.value

    #Display computer's choice
    print(f'Computer chose: {computer}\n')

    #Determine winner
    if player == computer:
      print("It's a tie!")
    elif wins[player] == computer:
      print(f'You won! {player} beats {computer}.')
    else:
      print(f'You lose! {computer} beats {player}.')

#Connect button to function
button.on_click(play_game)

#Display widgets
display(user, button, output)
```



## Number Guesser

```python
import random
import ipywidgets as widgets

from IPython.display import display

#Game settings
computer = random.randint(1,10)
max_attempts = 5
attempts = 0

print('Welcome to the Number Guesser Game!')
print('I picked a number between 1 and 10.')
print('You have 5 attempts to guess it.\n')

#List of numbers 1-10
numbers = list(range(1, 11))

#Create dropdown
user = widgets.Dropdown(
    options=numbers,
    value=1,
    description='Enter your guess: ',
    style = {'description_width': 'initial'},
    layout=widgets.Layout(width='250px')
)

#Create button
button = widgets.Button(
    description='Guess',
    button_style='success' #Makes button green
)

output = widgets.Output()

#Game function
def play_game(b):
  global attempts

  with output:
    output.clear_output()

  attempts += 1 #Count attempts
  guess = user.value

  print(f'\nAttempt: {attempts} of {max_attempts} | Your guess: {guess}')

  if guess == computer:
    print(f'Correct! The number was {computer}.')
    print(f'You guessed it in {attempts} attempt(s)!')
    button.disabled = True
  elif guess < computer:
    print('Too low! Try again.')
  else:
    print('Too high! Try again.')

  if attempts == max_attempts and guess != computer:
    print(f'\nGame over! The number was {computer}.')
    button.disabled = True

button.on_click(play_game)

display(user, button, output)
```
