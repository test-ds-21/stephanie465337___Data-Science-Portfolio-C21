# Hugging Face Data sets

**Hugging Face** has evolved far beyond a repository for transformer models; it is now the central hub for the open-source AI and machine learning ecosystem. For a data scientist, pushing a Parquet file to a Hugging Face Dataset repository turns a static file into a highly accessible, collaborative, and optimized ML asset.

Here is why it's a smart move for your workflow:

* **Seamless Pandas Integration:** Hugging Face integrates flawlessly with your existing workflow. Once your Parquet file is in an HF repo, you or your teammates can load it directly into a local Python environment using a single line of code (`load_dataset`), and convert it straight into a familiar **pandas DataFrame** with `.to_pandas()`.
* **Memory-Efficient Streaming:** Because Parquet is a columnar format and Hugging Face supports streaming, you can load and inspect massive datasets without running out of local RAM or waiting for giant files to fully download.
* **Version Control & Collaboration:** It essentially acts as "GitHub for data." You get full commit histories, automatic web previews of your data rows, and built-in discussion tabs, making it incredibly easy to share clean features or training sets across your team or the wider data science community.

- https://huggingface.co/settings/tokens/new?tokenType=fineGrained


### Save PAT and repo name in Colab Secrets Vault

- Create a secret named `hf_token`

- Create a secret named `hf_account`


## Authenticate to Hugging Face


```python
# Retrieve the token from Colab's secret manager
os.environ["HF_TOKEN"] = userdata.get('hf_token')
_ = os.environ["HF_TOKEN"]
f"{_[:5]} ... {_[-3:]}"
```

```python
os.environ["HF_ACCOUNT"] = userdata.get('hf_account')
hf_account = os.environ["HF_ACCOUNT"]
hf_account
```

```python
hf_repo = "Data_Science-21"
os.environ["HF_REPO"] = hf_repo
hf_repo
```

```python
# Log in automatically without an interactive prompt
!hf auth login --token $HF_TOKEN
```

## Push to Hugging Face


```python
%%capture hf_upload
%%bash

hf upload \
  --type dataset \
  ${HF_ACCOUNT}/${HF_REPO} \
  transactions.parquet
```

```python
print(hf_upload.stdout)
```

```python
%%capture hf_upload
%%bash

hf upload \
  --type dataset \
  ${HF_ACCOUNT}/${HF_REPO} \
  transactions.csv
```

```python
print(hf_upload.stdout)
```
