Find NAICS number in any NAICS column.

```python
cols = [ x for x in abq_bus.columns if "NAICS" in x and "Desc" not in x ]
filter = abq_bus[ cols ].apply( lambda row: '|'.join(row.values.astype(str)).find("541714") >= 0, axis=1)
abq_bus[ filter ]
```
