## Timeit

Numbers
- ms = 1/1e3 s = 1e-3 s = $10^{-3}$ s
- $\mu$s = 1/1e6 s = 1e-6 s = $10^{-6}$ s
- ns = 1/1e9 s = 1e-9 s = $10^{-9}$ s

````python
def sum(num):
  tally = 0
  for i in range(0,int(num+1)):
    tally += i
  return tally
````

````python
sum(100)
````

````python
def gaus_sum(num):
  return int(num*(num+1)/2)
````

````python
gaus_sum(100)
````

````python
max = 1e8
````

````python
%%timeit
sum(max)
````

````python
%%timeit
gaus_sum(max)
````
