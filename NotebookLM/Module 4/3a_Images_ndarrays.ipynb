# Image Representation of ndarrays

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
%config InlineBackend.figure_formats = 'retina'
```

## Display a numpy array of 0's and 1's.

```python
shades = np.array([0, 1, 0, 1, 0, 1, 0, 1, 0])

print(list(shades))
print(shades.shape)
print()
```

```python
#Convert from 1-D array to 2-D matrix
img = np.array([
    shades
])

print(img)
print(img.shape)
print()
```

```python
plt.imshow(img, cmap='gray');
```

## Using reshape

Go from 1-D to 2-D using the `reshape()` method.

```python
shades = np.array([0, 1, 0, 1, 0, 1, 0, 1, 0,])

print(list(shades))
print(shades.shape)
print()
```

```python
#Reshape into rows, columns (-1 == whatever it takes to work)
img = shades.reshape(1,-1)
print(img)
print(img.shape)
print()
```

```python
plt.imshow(img, cmap='gray');
```

```python
#Reshape into rows, columns (-1 == whatever it takes to work)
img = shades.reshape(3,-1)
print(img)
print(img.shape)
print()
```

```python
plt.imshow(img, cmap='gray');
```

## Using list repetition.

```python
#Repeating list [0, 1] using '*' list operator
shades = np.array([0, 1] * 4 + [0])

print(list(shades))
print(shades.shape)
print()
```

```python
#Reshape into rows, columns (-1 == whatever it takes to work)
img = shades.reshape(3,-1)
print(img)
print(img.shape)
print()
```

```python
plt.imshow(img, cmap='gray');
```

```python
#Repeating list [0, 1] using np.tile()
shades = np.tile([0, 1,], 5).flatten()[:9]

print(list(shades))
print(shades.shape)
print()
```

```python
#Reshape into rows, columns (-1 == whatever it takes to work)
img = shades.reshape(3, -1)
print(img)
print(img.shape)
print()
```

```python
plt.imshow(img, cmap='gray');
```

## Using `logical_not()` to invert an array


```python
shades = np.array(([ 0, 1] * 4 + [0]))

print(shades)
print(shades.shape)
print()
```

```python
#Reshape into rows, columns (-1 == whatever it takes to work)
img = np.logical_not(shades).reshape(3,-1)
print(img + 0)
print(img.shape)
print()
```

```python
plt.imshow(img, cmap='gray');
```

## Using `append()` to combine arrays


```python
shades = np.array(([0, 1] * 4 + [0]))

print(list(shades))
print(list(np.logical_not(shades) + 0 ))
print(shades.shape)
print()
```

```python
#Reshape into rows, columns (-1 == whatever it takes to work)
img = np.append(shades, np.logical_not(shades)).reshape(2,-1)
print(img)
print(img.shape)
print()
```

```python
plt.imshow(img, cmap='gray');
```

## Using modulo

```python
#Since the 0's and 1's alternate, this can be accomplished with a modulo (i.e. remainder) operator
shades = np.arange(18) % 2

print(list(shades))
print(shades.shape)
print()
```

```python
#Reshape into rows, columns (-1 == whatever it takes to work)
img = shades.reshape(2,-1)
print(img)
print(img.shape)
print()
```

```python
plt.imshow(img, cmap='gray');
```

## Create a diagonal line

```python
img = np.zeros((10, 10))
np.fill_diagonal(img, 1)

print(img)
print(img.shape)
print()
```

```python
plt.imshow(img, cmap='gray');
```

## Keep lower triangle

`tril()` keeps the lower triangle and fills the upper triangle with zeros.  By default, the diagonal is not included.  To include the diagonal, add the option `-1`.

```python
img = np.ones((10, 10))
img = np.tril(img)

print(img)
print(img.shape)
print()
```

```python
plt.imshow(img, cmap='gray');
```

## Keep upper triangle

`triu()` keeps the upper triangle and fills the lower triangle with zeros.  By default, the diagonal is not included.  To include the diagonal, add the option `-1`.

```python
img = np.ones((10, 10))
img = np.triu(img)

print(img)
print(img.shape)
print()

plt.imshow(img, cmap='gray');
```

## Keep the diagonal: `triu()` with `tril()`

The result is equivalent to using `fill_diagonal()`, but you can have a width.

```python
img = np.ones((10, 10))
img = np.triu(img, -1)
img = np.tril(img, 1)

print(img)
print(img.shape)
print()
```

```python
plt.imshow(img, cmap='gray');
```

## Checkerboard

```python
shape = (8, 8)
indices =  np.indices(shape)
print("== Indices: \n", indices)
print(indices.shape)
print()
```

```python
sum_of_i = indices.sum(axis = 0)
print("== Sum of the indices: \n", sum_of_i)
print(sum_of_i.shape)
print()
```

```python
checkerboard = sum_of_i % 2
img = checkerboard
print("== Checkerboard: \n", img)
print(img.shape)
print()
```

```python
plt.imshow(img, cmap='gray');
```

## Black and White numbers

Black is the minimum in the set of numbers.

White is the maximum in the set of numbers.

Notice how we use numpy's universal function feature to muliply all values in an array by a single value.

```python
shades = np.array([0, 1, 0, 1, 0, 1, 0, 1, 0,])

print(list(shades))
print(shades.shape)
print()
```

```python
#Convert from 1-D to 2-D
img = shades.reshape(1, -1)

print(img)
print(img.shape)
print()

plt.imshow(img, cmap='gray');
```

```python
#Multiply each value in matrix by a single value ( i.e. a scalar )
shades1 = (shades * 3.14) + 4

print(list(shades1))
print(shades1.shape)
print()
```

```python
#Convert from 1-D to 2-D
img = shades1.reshape(1, -1)

print(img)
print(img.shape)
print()

plt.imshow(img, cmap='gray');
```

```python
#Multiply each value in matrix by a single value (i.e. a scalar)
shades2 = (shades1 * -1)

print(list(shades2))
print(shades2.shape)
print()
```

```python
#Convert from 1-D to 2-D
img = shades2.reshape(1, -1)

print(img)
print(img.shape)
print()

plt.imshow(img, cmap='gray');
```

## Greyscale

Not everything has to be black and white.

```python
#Using modulo to generate numbers in addition to 0, 1.
n = 12
m = 4
shades = np.arange(n) % m
shades = (shades / (m-1)) * 255

print(list(shades))
print(shades.shape)
print()
```

```python
#Reshape into rows, columns (-1 == whatever it takes to work)
img = shades.reshape(1,-1)

print(list(img))
print(img.shape)
print()
```

```python
plt.imshow(img, cmap='gray');
```

Very fine grey scales.

```python
fig, sps = plt.subplots( 9, 1, figsize=(10,10))

for i, sp in enumerate(sps.flatten()):
  img = np.linspace( 0, 255, 2**(i+1) ).reshape(1,-1)
  sp.axis(False)
  sp.imshow(img, cmap='gray', aspect='auto')
```
